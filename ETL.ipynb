#### DESANIDAMOS LA COLUMNA cast

```python
data2['cast'] = data2['cast'].fillna('{}')
df1 = pd.json_normalize(data2['cast'].map(eval))
df1
```

```python
####VEMOS LAS CLAVES DE LA COLUMNA 'cast'
df1[2][0].keys()
```

```python
cast_id   = []
character = []
credit_id = []
gender    = []
id        = []
name      = []
order     = []
path      = []
for i in range(313):
    cid = 'cast_id' + str(i)
    char = 'character' + str(i)
    cr_id = 'credit_id' + str(i)
    gen = 'gender' + str(i)
    idd = 'id' + str(i)
    nam = 'name' + str(i)
    ord = 'order' + str(i)
    pat = 'profile_path' + str(i)
    cast_id.append(cid)
    character.append(char)
    credit_id.append(cr_id)
    gender.append(gen)
    id.append(idd)
    name.append(nam)
    order.append(ord)
    path.append(pat)
    df1[cid]   = pd.json_normalize(df1[i])['cast_id']
    df1[char]  = pd.json_normalize(df1[i])['character']
    df1[cr_id] = pd.json_normalize(df1[i])['credit_id']
    df1[gen]   = pd.json_normalize(df1[i])['gender']
    df1[idd]   = pd.json_normalize(df1[i])['id']
    df1[nam]   = pd.json_normalize(df1[i])['name']
    df1[ord]   = pd.json_normalize(df1[i])['order']
    df1[pat]   = pd.json_normalize(df1[i])['profile_path']
    
    

```

```python
col = list(range(313))
df1.drop(columns=col,inplace=True)
```

```python
df1['id_cast']       = df1[cast_id].apply(lambda x: ','   .join(x.dropna().astype(str)),axis=1)
df1['character']     = df1[character].apply(lambda x: ',' .join(x.dropna().astype(str)),axis=1)
df1['credit_id']     = df1[credit_id].apply(lambda x: ',' .join(x.dropna().astype(str)),axis=1)
df1['gender']        = df1[gender].apply(lambda x: ','    .join(x.dropna().astype(str)),axis=1)
df1['id']            = df1[id].apply(lambda x: ','        .join(x.dropna().astype(str)),axis=1)
df1['name']          = df1[name].apply(lambda x: ','      .join(x.dropna().astype(str)),axis=1)
df1['order']         = df1[order].apply(lambda x: ','     .join(x.dropna().astype(str)),axis=1)
df1['profilin_path'] = df1[path].apply(lambda x: ','      .join(x.dropna().astype(str)),axis=1)


```

```python
df1.drop(columns=cast_id, inplace=True)
```

```python
df1.drop(columns=character, inplace=True)
df1.drop(columns=credit_id, inplace=True)
df1.drop(columns=gender, inplace=True)
df1.drop(columns=id, inplace=True)
df1.drop(columns=name, inplace=True)
df1.drop(columns=order, inplace=True)
df1.drop(columns=path, inplace=True)
```
